```python
def getUrlTek(fileName):
    
    startName = 'https://www.tektorg.ru/procedures?name=&organizer%5B%5D=&datePublished_start=&datePublished_end=&page=1&sort=actual'

    cwd = os.getcwd()
    cwd+="\\Запросы\\"

    url_models_list = []
    searchTags = []
    organizers = []
    dateDiffer = 0
    with open(cwd+fileName,encoding='utf-8-sig') as file:
        for line in file:
            if line == 'Ключевые слова:\n':
                part=1
            elif line == 'Организатор(код из ТЭК-Торг):\n':
                part=2
            elif line == 'Период поиска(дни):\n':
                part=3
            else:
                if line!="\n":
                    if part==1:
                        searchTags.append(line)
                    elif part==2:
                        organizers.append(line)
                    else:
                        dateDiffer = int(line)

    if searchTags == []:
        nameToAdd = startName
        zzz = (f"{(datetime.now()- timedelta(days=dateDiffer)):%d.%m.%Y}")
        zzz2 = (f"{(datetime.now()):%d.%m.%Y}")
        zzz3 = ""
        for ppp in organizers:
            zzz4 = ppp.replace('/',"%2F")
            zzz3+=zzz4[:-1]+"%3B"
        zzz3 = zzz3[:-3]
        nameToAdd = (nameToAdd.replace('name=?',''))
        if organizers == []:
            nameToAdd = (nameToAdd.replace('&organizer%5B%5D=',''))
        else:
            nameToAdd = (nameToAdd.replace('organizer%5B%5D=','organizer%5B%5D='+zzz3))
        nameToAdd = (nameToAdd.replace('datePublished_start=','datePublished_start='+zzz))
        nameToAdd = (nameToAdd.replace('datePublished_end=','datePublished_end='+zzz2))
        url_models_list.append(nameToAdd)
    else:
        for tag in searchTags:
            nameToAdd = startName
            zzz = (f"{(datetime.now()- timedelta(days=dateDiffer)):%d.%m.%Y}")
            zzz2 = (f"{(datetime.now()):%d.%m.%Y}")
            zzz3 = ""
            for ppp in organizers:
                zzz4 = ppp.replace('/',"%2F")
                zzz3+=zzz4[:-1]+"%3B"
            zzz3 = zzz3[:-3]
            tagToAdd = tag[:-1].replace(" ","+")
            nameToAdd = (nameToAdd.replace('name=','name='+tagToAdd))
            if organizers == []:
                nameToAdd = (nameToAdd.replace('&organizer%5B%5D=',''))
            else:
                nameToAdd = (nameToAdd.replace('organizer%5B%5D=','organizer%5B%5D='+zzz3))
            nameToAdd = (nameToAdd.replace('datePublished_start=','datePublished_start='+zzz))
            nameToAdd = (nameToAdd.replace('datePublished_end=','datePublished_end='+zzz2))
            url_models_list.append(nameToAdd)

    return(url_models_list)   
```

```python
def numPageTek(adress):
    chrome_options = Options()
    chrome_options.page_load_strategy = 'normal'
    chrome_options.add_argument("--disable-extensions")
#     chrome_options.add_argument('--headless')
    chrome_options.add_argument('--log-level=3')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--ignore-certificate-errors')
    chrome_options.add_argument('--ignore-certificate-errors-spki-list')
    chrome_options.add_argument('--ignore-ssl-errors')
    chrome_options.add_argument('--disable-infobars')
    chrome_options.add_experimental_option('excludeSwitches', ['enable-logging'])
    driver = webdriver.Chrome(options=chrome_options)
#     driver = webdriver.Chrome()
#     driver.implicitly_wait(waitTimeTek)
    driver.get(adress)
    time.sleep(waitTimeTek)
#     driver.implicitly_wait(waitTimeTek)
    html = driver.page_source
    page = BeautifulSoup(html, 'lxml')
    
    total =  page.find_all(class_="ProceduresSortstyled__Total-sc-1pzpmm5-1")
    driver.close()
    
    zz = str(total)
    
    b = (zz.find('<span>'))
    c = (zz.find('</span>'))
    v = (int(zz[b+6:c]))
    if v==0:
        return 0
    else:
        return (v//10)+1
```

```python
def pageContentTek(adress):
    
    pageStart='1'

    chrome_options = Options()
    chrome_options.page_load_strategy = 'normal'
    chrome_options.add_argument("--disable-extensions")
#     chrome_options.add_argument('--headless')
    chrome_options.add_argument('--log-level=3')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--ignore-certificate-errors')
    chrome_options.add_argument('--ignore-certificate-errors-spki-list')
    chrome_options.add_argument('--ignore-ssl-errors')
    chrome_options.add_argument('--disable-infobars')
    chrome_options.add_experimental_option('excludeSwitches', ['enable-logging'])
    
    print("Обрабатывается запрос:" + adress)
    print()
    
    totalPage = numPageTek(adress)
    
    print("По запросу будет обработано: "+str(totalPage)+" стр.")
    
    dfOut = pd.DataFrame(columns=['organizer','name','number','type','dateStart',
                                  'dateFinish','nmc','url'])
    
    for i in range(int(pageStart),int(totalPage)+1):
        page=1
        nameToParse = adress.replace("page=1","page="+str(i))
        print("Обрабатывается - "+str(i)+"-ая стр.")

        driver = webdriver.Chrome(options=chrome_options)
        driver.implicitly_wait(waitTimeTek)
        driver.get(nameToParse)
        time.sleep(waitTimeTek)
        driver.implicitly_wait(waitTimeTek)
        
        class_id = SoupStrainer(attrs={'class': re.compile('^CardProcedureViewstyled__Container-sc-1f5zioq-0.*')})
        html = driver.page_source
        page = BeautifulSoup(html, 'lxml', parse_only=class_id)

        allCards = page.find_all(class_="CardProcedureViewstyled__Container-sc-1f5zioq-0")

        for card in allCards:
            total5 = card.find_all(class_="ProcedureInfoHeaderstyled__RegistryNumber-sc-zou3o8-2")
            total = card.find_all(class_="CardProcedureViewstyled__Title-sc-1f5zioq-7")
            total2 = card.find_all(class_="CardProcedureViewstyled__BottomTopValue-sc-1f5zioq-10")
            total3 = card.find_all(class_="CardProcedureViewstyled__BottomLeftValue-sc-1f5zioq-14")
            total6 = card.find_all(class_='ProcedurePricestyled__Container-sc-h9qmqw-0')
            total4 = card.find_all(class_="DateTimeGMTstyled__DTGDatetime-sc-qm732q-0")
            
            dfOut1 = pd.DataFrame(index=range(1),columns=['organizer','name','number','type','dateStart',
                                  'dateFinish','nmc','url'])       

        # Достаем url и название
            urlname=str(total[0])
            b = (urlname.find('href="'))
            c = (urlname.find('>'))
            c2 = (urlname.find('</a>'))
            dfOut1.url =str("https://www.tektorg.ru"+urlname[b+6:c-1])
            nnn = str(urlname[c+1:c2]).replace('\n',' ')
            dfOut1.name = nnn

        # Достаем номер
            numb = str(total5[0])
            b = (numb.find(">"))
            c = (numb.find("</span>"))
            dfOut1.number =str(numb[b+3:c])
            
        # Достаем НМЦ
            nmc = str(total6[0])
            b = (nmc.find(">"))
            c = (nmc.find("</div>"))
            dfOut1.nmc =str(nmc[b+1:c])
            
        # Достаем организатора
            organiz = str(total2[0])
            b = (organiz.find(">"))
            c = (organiz.find("</div>"))
            dfOut1.organizer =str(organiz[b+1:c])

        # Достаем статус
            status = str(total3[0])
            b = (status.find(">"))
            c = (status.find("</div>"))
            dfOut1.type =str(status[b+1:c])
            
        # Достаем дату начала
            if len(total4)>(1):
                dataStar = str(total4[0])
                b = (dataStar.find("datetime="))
                c = (dataStar.find("><span"))
                ttt = str(dataStar[b+10:c-1])
                ttt = ttt.replace("T"," ")
                ttt = ttt.replace("-",".")
                dfOut1.dateStart =ttt[8:10]+ttt[4:8]+ttt[0:4]+ttt[11:16]+"GMT+"+ttt[-4:-3]

        # Достаем дату окончания
            if len(total4)>(1):
                dataFin = str(total4[1])
                b = (dataFin.find("datetime="))
                c = (dataFin.find("><span"))
                ttt= str(dataFin[b+10:c-1])
                ttt = ttt.replace("T"," ")
                ttt = ttt.replace("-",".")
                dfOut1.dateFinish =ttt[8:10]+ttt[4:8]+ttt[0:4]+ttt[11:16]+"GMT+"+ttt[-4:-3]

        # Добавляем в общий массив
            dfOut1.fillna("", inplace=True)
            dfOut = pd.concat([dfOut,dfOut1],ignore_index=True, axis=0)

    if 'driver' in locals():
        driver.close()
    
    print("Обработка запроса по данному адресу завершена")
    print()
    return(dfOut)
```

```python
def pageContentGPB(adress):
    
    pageStart='1'

    chrome_options = Options()
    chrome_options.page_load_strategy = 'normal'
    chrome_options.add_argument("--disable-extensions")
    # chrome_options.add_argument('--headless')
    chrome_options.add_argument('--log-level=3')
    chrome_options.add_argument('--ignore-certificate-errors')
    chrome_options.add_argument('--ignore-certificate-errors-spki-list')
    chrome_options.add_argument('--ignore-ssl-errors')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--disable-infobars')
    chrome_options.add_experimental_option('excludeSwitches', ['enable-logging'])

    print("Обрабатывается запрос:" + adress)
    print()

    totalPage = numPageGPB(adress)

    print("По запросу будет обработано: "+str(totalPage)+" стр.")

    dfOut = pd.DataFrame(columns=['organizer','name','number','type','dateStart',
                                  'dateFinish','nmc','url'])


    for i in range(int(pageStart),int(totalPage)+1):
        page=1
        nameToParse = adress.replace("page=1","page="+str(i))
        print("Обрабатывается - "+str(i)+"-ая стр.")

        driver = webdriver.Chrome(options=chrome_options)
        driver.implicitly_wait(waitTimeGPB)
        driver.get(nameToParse)
        time.sleep(waitTimeGPB)
        driver.implicitly_wait(waitTimeGPB)

        class_id = SoupStrainer(attrs={'class': re.compile('^globalProceduresCard.*')})
        
        html = driver.page_source
        page = BeautifulSoup(html, 'lxml', parse_only=class_id)

        allCards = page.find_all(class_="globalProceduresCard")

        for card in allCards:
            total = card.find_all(class_="globalProceduresCard__title link link--blackToLight")
            total5 = card.find_all(class_="link link--toSteelBlue")
            total2 = card.find_all(class_="globalProceduresCard__organizerValue")
            total3 = card.find_all(class_="globalProceduresCard__status")
            total6 = card.find_all(class_='globalProceduresCard__priceValue')
            total4 = card.find_all(class_="globalProceduresCard__expiryDateValueDate")

            dfOut1 = pd.DataFrame(index=range(1),columns=['organizer','name','number','type','dateStart',
                                  'dateFinish','nmc','url'])    
            
        # Достаем url и название'
            if total!=[]:
                urlname=str(total[0])
                b = (urlname.find('href="'))
                c = (urlname.find('target='))
                
                if urlname[b+6]=="h":
                    dfOut1.url =str(urlname[b+6:c-2])
                else:
                    dfOut1.url =str("https://www.etpgpb.ru"+urlname[b+6:c-2])
                pp=""
                flag = 0
                v = (urlname.find('_self'))
                if v>0:
                    for i in urlname[v+7:]:
                        if i=="<" and flag ==0:
                            flag=1
                        elif i==">" and flag==1:
                            flag=0
                        else:
                            if flag==0:pp+=i
                else:
                    v = (urlname.find('_blank'))
                    if v>0:
                        for i in urlname[v+8:]:
                            if i=="<" and flag ==0:
                                flag=1
                            elif i==">" and flag==1:
                                flag=0
                            else:
                                if flag==0:pp+=i
                dfOut1.name = pp


        # Достаем номер
            if total5!=[]:
                numb = str(total5[0])
                b = (numb.find(">"))
                c = (numb.find("</a>"))
                dfOut1.number =str(numb[b+3:c])

        # Достаем НМЦ
            if total6!=[]:
                nmc = str(total6[0])
                b = (nmc.find("Integer"))
                if b>0:
                    cc = nmc.find("</span>")
                    pp = str(nmc[b+9:cc])
                    dfOut1.nmc =pp.replace(' ','')
                else:
                    dfOut1.nmc ="Цена не указана"

        # Достаем организатора
            if total2!=[]:
                organiz = str(total2[0])
                b = (organiz.find("<span>"))
                c = (organiz.find("</span>"))
                pp = str(organiz[b+6:c])

                dfOut1.organizer =pp

        # Достаем статус
            if total3!=[]:
                status = str(total3[0])
                b = (status.find(">"))
                c = (status.find("</div>"))
                ppp = str(status[b+1:c]).replace("\n",'')
                zzz = ppp.split(' ')
                ret = ""
                for ast in zzz:
                    if ast!="":
                        ret= ret + ast+" "
                ret = ret[:-1]
                dfOut1.type =ret

        # Достаем дату окончания
            if total4!=[]:
                dateFin = str(total4)
                b = (dateFin.find(">"))
                c = (dateFin.find("</span>"))
                ttt= str(dateFin[b+1:c-1]).replace('\n','')
                if "Сегодня" in ttt:
                    ppp = ttt.replace("Сегодня",f"{(datetime.now()):%d.%m.%Y}")
                    fff = ppp.replace(", ","")+"GMT+3"
                    dfOut1.dateFinish = fff.replace(" ","")
                elif "Завтра" in ttt:
                    ppp = ttt.replace("Завтра",(f"{(datetime.now()+ timedelta(days=1)):%d.%m.%Y}"))
                    fff = ppp.replace(", ","")+"GMT+3"
                    dfOut1.dateFinish = fff.replace(" ","")
                elif "Вчера" in ttt:
                    ppp = ttt.replace("Вчера",(f"{(datetime.now()- timedelta(days=1)):%d.%m.%Y}"))
                    fff = ppp.replace(", ","")+"GMT+3"
                    dfOut1.dateFinish = fff.replace(" ","")  
                else:
                    fff = ttt.replace(", ","")+"GMT+3"
                    dfOut1.dateFinish = fff.replace(" ","")

        # Добавляем в общий массив
            dfOut1.fillna("", inplace=True)
            dfOut = pd.concat([dfOut,dfOut1],ignore_index=True, axis=0)
    
    if 'driver' in locals():
        driver.close()
    print("Обработка запроса по данному адресу завершена")
    print()       
    return(dfOut)
```

```python
#Перебор по одной странице лист адресов
cwd = os.getcwd()
cwd+="\\Запросы\\"

# parent_dir = "Z:\\"
# parent_dir = "C:\\Users\\DZ\\Programs\\"
parent_dir = "C:\\"

directory =  str(f"{(datetime.now()):%d-%m-%Y}")
 
dirName = os.path.join(parent_dir, directory)
if not os.path.exists(dirName):
    os.mkdir(dirName)

for fileName in os.listdir(cwd):
    url_models_list_Tek = getUrlTek(fileName)
    url_models_list_GPB = getUrlGPB(fileName)
    for jjj in range(len(url_models_list_Tek)):

        dfOutTek = pageContentTek(url_models_list_Tek[jjj])
        dfOutGPB = pageContentGPB(url_models_list_GPB[jjj])
        dfOut = pd.concat([dfOutTek,dfOutGPB], ignore_index=True, axis=0)
        
        aaa = url_models_list_Tek[jjj].find("name=")
        bbb = url_models_list_Tek[jjj].find("&")
        fff = url_models_list_Tek[jjj][aaa+5:bbb].replace("+",' ')
        uuu = fileName.replace('.txt','')
        if fff=="":
            nameForOut =uuu + " "
        else:
            nameForOut =uuu +"+"+ fff + " "
        nameForOut+=str(f"{(datetime.now()):%d-%m-%Y}")+'.xlsx'
        
        if not dfOut.empty:
            print("Запрос "+url_models_list_Tek[jjj]+" выгружен в файл "+nameForOut)
            print()
            dfOut.to_excel((os.path.join(dirName, nameForOut)),
#                            encoding='utf-8-sig',
                           index=True,
                           header = ['Заказчик (Доч.общ.)','Наименование процедуры (ссылка)','Номер процедуры',
                                     'Опрос рынка/Закупка','Дата публикации','Срок окончания подачи заявок',
                                    'НМЦ','Ссылка'])
```
